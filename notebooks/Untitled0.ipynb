```python
*
This code is used to generate numbers for ACCJC Annual Report.
The code was developed in May 05, 2020;
Updated: March 2021

*/

--** Total unduplicated enrollment

-- Unduplicated Enrollment
select 'FY 19/20' as AcademicYear,  count(distinct id) as UnduplicatedCount
from BAN_STUDENT_ACADEMIC_HISTORY h
where h.term_code in ('201830','201910','201920') and substring(h.college_code, 1, 1)='2' and h.final_grade <>'UD'
union
select 'FY 18/19', count(distinct id) "FY 18/19"
from BAN_STUDENT_ACADEMIC_HISTORY h
where h.term_code in ('201730','201810','201820') and substring(h.college_code, 1, 1)='2' and h.final_grade <>'UD'
union
select 'FY 17/18', count(distinct id) "FY 17/18"
from BAN_STUDENT_ACADEMIC_HISTORY h
where h.term_code in ('201630','201710','201720') and substring(h.college_code, 1, 1)='2' and h.final_grade <>'UD'



-------------------------------------------** Total unduplicated enrollment in degree applicable credit courses
;
-- FY19/20
SELECT 'FY 19/20' as AcademicYear, count(distinct SAH.id) as UnduplicatedHeadcount
FROM BAN_STUDENT_ACADEMIC_HISTORY AS SAH
LEFT JOIN [IRP].[dbo].[MIS_CB] AS CB ON ( SAH.course = LEFT(CB.CB01,PATINDEX('%[0-9]%',CB.CB01)-1) + ' ' + RTRIM(SUBSTRING(CB.CB01,PATINDEX('%[0-9]%',CB.CB01),LEN(CB.CB01))) )
WHERE
       substring(sah.college_code, 1, 1)='2' and
	   SAH.term_code in ('201830','201910','201920')
	   and SAH.final_grade <>'UD'
	   AND CB.CB04 = 'D'
	   union
-- FY18/19
SELECT 'FY 18/19' as AcademicYear, count(distinct SAH.id) as UnduplicatedHeadcount
FROM BAN_STUDENT_ACADEMIC_HISTORY AS SAH
LEFT JOIN [IRP].[dbo].[MIS_CB] AS CB ON ( SAH.course = LEFT(CB.CB01,PATINDEX('%[0-9]%',CB.CB01)-1) + ' ' + RTRIM(SUBSTRING(CB.CB01,PATINDEX('%[0-9]%',CB.CB01),LEN(CB.CB01))) )
WHERE
       substring(sah.college_code, 1, 1)='2' and
	   SAH.term_code in ('201730','201810','201820')
	   and SAH.final_grade <>'UD'
	   AND CB.CB04 = 'D'
union
--FY17/18
SELECT 'FY 17/18' as AcademicYear, count(distinct SAH.id) "FY 17/18"
FROM BAN_STUDENT_ACADEMIC_HISTORY as SAH
LEFT JOIN [IRP].[dbo].[MIS_CB] AS CB ON ( SAH.course = LEFT(CB.CB01,PATINDEX('%[0-9]%',CB.CB01)-1) + ' ' + RTRIM(SUBSTRING(CB.CB01,PATINDEX('%[0-9]%',CB.CB01),LEN(CB.CB01))) )
WHERE
       substring(sah.college_code, 1, 1)='2' and
	   SAH.term_code in ('201630','201710','201720')
	   and SAH.final_grade <>'UD'
	   AND CB.CB04 = 'D'
;
-------------------------------------------** Total unduplicated enrollment in distance education

-- FY19/20
select 'FY 19/20' as AcademicYear,
count(distinct h.id) as UnduplicatedHeadcount
from BAN_STUDENT_ACADEMIC_HISTORY h
	join BAN_TERM_SECTION_FTE f
	on f.term_code=h.term_code and f.crn=h.crn
where h.term_code in ('201830','201910','201920') and
substring(h.college_code, 1, 1)='2' and
h.final_grade <>'UD' and
f.method_of_instr_code = '72'
union
-- FY18/19
select 'FY 18/19' as AcademicYear,
count(distinct h.id) "FY 18/19"
from BAN_STUDENT_ACADEMIC_HISTORY h
	join BAN_TERM_SECTION_FTE f
	on f.term_code=h.term_code and f.crn=h.crn
where h.term_code in ('201730','201810','201820') and
substring(h.college_code, 1, 1)='2' and
h.final_grade <>'UD' and
f.method_of_instr_code ='72'
union
--FY17/18
select 'FY 17/18' as AcademicYear,
count(distinct h.id) "FY 17/18"
from BAN_STUDENT_ACADEMIC_HISTORY h
	join BAN_TERM_SECTION_FTE f
	on f.term_code=h.term_code and f.crn=h.crn
where h.term_code in ('201630','201710','201720') and
substring(h.college_code, 1, 1)='2' and
h.final_grade <>'UD' and
f.method_of_instr_code = '72'
;

------ Course Completion (SUCCESS) Rate


select 'FY 17/18',
round(cast(count(case when h.final_grade in ('A','B','C','CR','P','INB','INC') then pidm end ) as float) / cast(count(*) as float) * 100.0, 1) as succes
from BAN_STUDENT_ACADEMIC_HISTORY h
where h.term_code in ('201630','201710','201720') and substring(h.college_code, 1, 1)='2'
and h.final_grade not in ( 'DR', 'IP', 'IX', 'MW', 'RD', 'UD', 'XX')
union
select 'FY 18/19',
round(cast(count(case when h.final_grade in ('A','B','C','CR','P','INB','INC') then pidm end ) as float) / cast(count(*) as float) * 100.0, 1) as succes
from BAN_STUDENT_ACADEMIC_HISTORY h
where h.term_code in ('201730','201810','201820') and substring(h.college_code, 1, 1)='2'
and h.final_grade not in ( 'DR', 'IP', 'IX', 'MW', 'RD', 'UD', 'XX')
union
select 'FY 19/20' as AcademicYear, --final_grade,
round(cast(count(case when h.final_grade in ('A','B','C','CR','P','INB','INC') then pidm end ) as float) / cast(count(*) as float) * 100.0, 1) as success
from BAN_STUDENT_ACADEMIC_HISTORY h
where h.term_code in ('201830','201910','201920') and substring(h.college_code, 1, 1)='2'
and h.final_grade not in ( 'DR', 'IP', 'IX', 'MW', 'RD', 'UD', 'XX')
--group by h.final_grade
;

------------- Certificates (approved 16 units or more)

select 'FY 17/18', count(distinct pidm)
from [dbo].[BAN_TERM_DEGREES_CERTIFICATES] d
where d.term_code in ('201630','201710','201720')
and d.degree_certif_status_code in ('AW','PN')
and d.degree_certif_code in ('N','L','T','F')
and d.college_code like '2%'
union
select 'FY 18/19', count(distinct pidm)
from [dbo].[BAN_TERM_DEGREES_CERTIFICATES] d
where d.term_code in ('201730','201810','201820')
and d.degree_certif_status_code in ('AW','PN')
and d.degree_certif_code in ('N','L','T', 'F')
and d.college_code like '2%'
union
select 'FY 19/20', count(distinct pidm)
from [dbo].[BAN_TERM_DEGREES_CERTIFICATES] d
where d.term_code in ('201830','201910','201920')
and d.degree_certif_status_code in ('AW','PN')
and d.degree_certif_code in ('N','L','T', 'F')
and d.college_code like '2%'

------------- Degrees

select 'FY 17/18'--, d.degree_certif_code,d.degree_certif_desc
, count(distinct pidm) distinct_students, count(*) degrees
from [dbo].[BAN_TERM_DEGREES_CERTIFICATES] d
where d.term_code in ('201630','201710','201720')
and d.degree_certif_status_code in ('AW','PN')
and d.degree_certif_code in ('A','AA-T','AS-T', 'S')
and d.college_code like '2%'
--group by degree_certif_code, d.degree_certif_desc
---order by 2
union
select 'FY 18/19', count(distinct pidm) distinct_students, count(*) degrees
from [dbo].[BAN_TERM_DEGREES_CERTIFICATES] d
where d.term_code in ('201730','201810','201820')
and d.degree_certif_status_code in ('AW','PN')
and d.degree_certif_code in ('A','AA-T','AS-T', 'S')
and d.college_code like '2%'
union
select 'FY 19/20', count(distinct pidm) distinct_students, count(*) degrees
from [dbo].[BAN_TERM_DEGREES_CERTIFICATES] d
where d.term_code in ('201830','201910','201920')
and d.degree_certif_status_code in ('AW','PN')
and d.degree_certif_code in ('A','AA-T','AS-T', 'S')
and d.college_code like '2%'
;
```
