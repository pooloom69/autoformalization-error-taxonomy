```python
import os
from collections import Counter

label_dir = "/content/drive/MyDrive/SmartCookingAssistant/valid/labels"  # 네 경로로 수정
class_names = [
    "empty_burner",
    "empty_burner_on",
    "empty_pan",
    "empty_pot",
    "empty_with_water",
    "pan_with_oil",
    "pot_water_boiling",
    "pot_with_water",
]

counter = Counter()

for fname in os.listdir(label_dir):
    if not fname.endswith(".txt"):
        continue
    with open(os.path.join(label_dir, fname), "r") as f:
        for line in f:
            if line.strip():
                class_id = int(line.split()[0])
                counter[class_id] += 1

for i, name in enumerate(class_names):
    print(i, name, counter[i])
```
